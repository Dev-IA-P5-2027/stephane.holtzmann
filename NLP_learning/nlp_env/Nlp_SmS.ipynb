```python
import pandas as pd
import re
import string
import nltk

nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")
```

```python
df = pd.read_csv("SMS_test.csv", encoding="latin-1")
df.head(3)
```

```python
def text_lowercase(text):
    return str(text).lower()

df["Case Normalization"] = df["Message_body"].apply(text_lowercase)
df.head(3)
```

```python
import contractions

def fix_contractions(text):
    return contractions.fix(str(text))

df["Case Normalization"] = df["Case Normalization"].apply(fix_contractions)
df.head(3)
```

```python
def remove_urls(text):
    return re.sub(r"https?://\S+|www\.\S+", "", str(text))

df["Noise Removal"] = df["Case Normalization"].apply(remove_urls)
df.head(3)
```
