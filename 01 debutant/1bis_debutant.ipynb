Exercice 6 :
Écrire un programme en python pour réviser ses tables de multiplication. Le programme tire au hasard deux entiers entre 0 et 10 et demande à l'utilisateur de donner leur produit.
Si l'utilisateur répond juste, il gagne un point. Si il répond faux, il perd un point.
On interrogera 5 fois l'utilisateur.
Le programme affiche à la fin le score de l'utilisateur.

```python
import random

def quiz_multiplication(nb_questions: int = 5) -> int:
    """
    Pose nb_questions de multiplication aléatoires à l'utilisateur.
    Donne +1 point pour chaque bonne réponse, -1 sinon.
    Retourne le score final.
    """
    score = 0

    for i in range(nb_questions):
        a = random.randint(0, 10)
        b = random.randint(0, 10)
        print(f"\nQuestion {i+1} : {a} x {b} = ?")

        try:
            reponse = int(input("Ta réponse : "))
        except ValueError:
            print("Entrée invalide — 0 point cette fois.")
            continue  # passe à la question suivante

        if reponse == a * b:
            print("Juste! +1 point")
            score += 1
        else:
            print(f"Faux ! La bonne réponse était {a * b}. -1 point")
            score -= 1

        print(f"Score actuel : {score}")

    return score


# --- Partie principale ---
print("révision des tables de multiplication !")

score_final = quiz_multiplication(5)
print(f"\n,Ton score final est : {score_final}/5")
```

L'ordinateur tire un nombre entier au hasard entre 0 et 100.
L'utilisateur doit le trouver et pour cela propose des valeurs.
L'ordinateur indique pour chaque valeur proposée si la valeur est trop petite, trop grande ou s'il a trouvé.
a) Écrire un programme en Python pour jouer à ce jeu.
b) Modifier le programme pour qu'il s'arrête si l'utilisateur n'a pas trouvé au bout de 10 coups.
c) Mettre ce programme dans une fonction plus_ou_moins() qui prendra en paramètres :

    le nombre min et max de la plage de tirage du nombre au hasard
    le nombre de coups maximum


```python
import random

def plus_ou_moins(min_val: int, max_val: int, max_coups: int):
    """
    Jeu du 'plus ou moins'.
    L'ordinateur tire un nombre entre min_val et max_val.
    L'utilisateur doit le deviner en un nombre limité de coups.
    """
    nombre_secret = random.randint(min_val, max_val)
    coups = 0

    print(f" Devine le nombre mystère entre {min_val} et {max_val} !")

    while coups < max_coups:
        try:
            proposition = int(input(f"Essai {coups+1}/{max_coups} → Ta proposition : "))
        except ValueError:
            print(" Merci d'entrer un nombre entier.")
            continue

        coups += 1

        if proposition < nombre_secret:
            print(" Trop petit !")
        elif proposition > nombre_secret:
            print("Trop grand !")
        else:
            print(f"✅Bravo ! Tu as trouvé en {coups} coups.")
            return True  # partie gagnée

    print(f"Tu as épuisé tes {max_coups} essais. Le nombre était {nombre_secret}.")
    return False  # partie perdue


# --- Partie principale ---
print("=== Jeu : Devine le nombre ===")

# paramètres : borne min, borne max, nombre d'essais maximum
plus_ou_moins(0, 100, 10)

```

Ecrire un programme qui tourne indéfiniement et qui demande à l'utilisateur de rentrer un nombre entier compris dans une plage définie aléatoirement (par exemple, entrer un nombre compris entre 12 et 34).

```python
def saisie_notes():
    print("Saisie des notes. Tapez 'fin' pour terminer.")
    notes = []
    while True:
        entré= input("entrer note sur 20 : ")
        if entré=='':
            break
        try:
            note = float(entré) 
        except ValueError:
            print("Entrée invalide")
            continue
        if 0 <= note <= 20:
            notes.append(note)
        else:
            print("entre 0 et 20 svp")
    if len(notes) == 0:
        print("aucune note saisie.)")
        return 0, None, None, None
    nbre = len(notes)
    note_max = max(notes)
    note_min = min(notes)
    moyenne = sum(notes) / nbre
    return nbre , note_max, note_min, moyenne


saisie_notes()
```
