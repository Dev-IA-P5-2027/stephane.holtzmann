# Workshop 2 – FP-Growth pour extraire des règles d'association

```python


import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rule

transactions = [
    ['bread', 'milk'],
    ['bread', 'diaper', 'beer', 'eggs'],
    ['milk', 'diaper', 'beer', 'cola'],
    ['bread', 'butter'],
    ['bread', 'milk', 'diaper', 'beer'],
    ['bread', 'milk', 'diaper', 'cola'],
    ['milk', 'eggs'],
    ['beer', 'diaper'],
    ['bread', 'beer'],
    ['milk', 'diaper', 'eggs'],
]


te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

print("Aperçu du tableau binaire :")
print(df.head())


frequent_itemsets_fp = fpgrowth(df, min_support=0.3, use_colnames=True)
print("\nItemsets fréquents (FP-Growth) :")
print(frequent_itemsets_fp)


rules_fp = association_rules(frequent_itemsets_fp, metric="lift", min_threshold=1.0)
rules_fp_sorted = rules_fp.sort_values(by="lift", ascending=False)

print("\nRègles d'association (FP-Growth) triées par lift :")
print(rules_fp_sorted[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```
